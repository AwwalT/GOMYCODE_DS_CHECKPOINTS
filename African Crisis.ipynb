```python
def skew_function():
    
    """ I want to pull out all the skewed columns in the dataframe"""
    
    skew_value = []
    
    for col in merged.select_dtypes('number'):
        skew_col = merged[col].skew()
        skew_value.append((col , skew_col))
    return sorted(skew_value , key = lambda x : x[1] , reverse = True)
```

```python
def outliers_check():
        cols_store = []
    
        for col in merged.select_dtypes("number"): 
            q1 = merged[col].quantile(.25)
            q3 = merged[col].quantile(.75)
            iqr = q3 - q1
            lower_out= (merged[col] < (q1 - (1.5*iqr))).sum()
            upper_out = (merged[col]  > (q3 + (1.5*iqr))).sum()

            if lower_out > 0 or upper_out > 0 :
                cols_store.append(col)

        return cols_store
```
