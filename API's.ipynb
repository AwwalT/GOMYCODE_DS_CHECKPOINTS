## API
Stands for Application programming interface: it allows different software that are built differently to communicate with one another. Its a set of protocols and routines for buildings and interacting with software applications, a standard form of transferring data through API is JSON(Javascript Object Notation).

Work in a server client mode
- Json is the end product of the request
- REST: This is the means of communication, done through the web
- GET: used to request data
- POST: used to write data to the server(new data)
- DELETE: delete data
- PUT: write(update data)
